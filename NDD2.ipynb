```python
data.rename(columns = {'Are you a Nigerian?':'Nationality', 
                    'Are you married?':'Married', 
                    'Do you have kids?':'Kids', 
                    'Where do you live?':'Location',
                    'What part of your state do you live in?':'Location Status',
                    'Are you Happy with your job?':'Job Satisfaction',
                    'Is this your primary source of income?':'Income Source',
                    'What is your steady/constant income range in a month?':'Income Range', 
                    'Do you think it is enough for you and your family?':'Income Satisfaction',
                    'What are your other sources of income? ':'Other Income Sources', 
                    'Do you pay your tax?':'Tax', 
                    'What is your largest form of expense':'Large Expense Form', 
                    'If there was one thing you really wanted to own, what would that be?':'Wish List',
                    'How much do you think you earn from all other sources of income in a month':'Total Income Accrued',
                    'On an average, how much do you think you spend in a month?':'Average Expenses'}, inplace = True)

data.head(3)
```

```python
df = data
df
```

```python
df.dropna()
```

```python
#Convert the string columns into integer columns by converting the words into numbers
df['Nationality'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df['Gender'].replace(to_replace = ['Male', 'Female'], value = [0,1], inplace = True)
df['Location Status'].replace(to_replace = ['Rural', 'Urban', 'City'], value = [0,1,2], inplace = True)
df['Kids'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df['Married'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df['Job Satisfaction'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df['Income Source'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df['Income Satisfaction'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df['Tax'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df['Do you think Nigeria can get better?'].replace(to_replace = ['Yes', 'No'], value = [0,1], inplace = True)
df.head(3)
```
